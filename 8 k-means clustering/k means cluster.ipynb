> **Problem :** Apply the K-Means clustering algorithm to cluster the Iris dataset into different groups based on the features, and then evaluate the results.

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.sample(10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['sepal_length'],df['petal_width'])
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:,[0,3]].values
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,random_state=42)
y_means = km.fit_predict(x)
y_means
```

```python
plt.scatter(x[y_means == 0,0],x[y_means == 0,1],color='red',label='~ setosa')
plt.scatter(x[y_means == 1,0],x[y_means == 1,1],color='blue',label='~ versicolor')
plt.scatter(x[y_means == 2,0],x[y_means == 2,1],color='green',label='~ virginica')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.title('K-Means Clustering of Iris Dataset')
```
