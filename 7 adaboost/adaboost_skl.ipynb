> **Problem :** Implement the AdaBoost algorithm using Scikit-Learn for a classification task and observe how combining weak learners improves the model's accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
np.random.seed(42)
x,y = make_circles(n_samples=500,factor=0.1,noise=0.4,random_state=42) #data
plt.scatter(x[:,0],x[:,1], c=y)
```

```python
def plot_decision_boundry(clf):
    x_range = np.linspace(x.min(),x.max(),100)
    xx1,xx2 = np.meshgrid(x_range,x_range)
    y_hat = clf.predict(np.c_[xx1.ravel(),xx2.ravel()])
    y_hat= y_hat.reshape(xx1.shape)
    plt.contour(xx1,xx2,y_hat,alpha=0.2)
    plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis',alpha=.7)
    plt.title('adaboost clf')
    plt.show()
```

> **only 1 weak learner**

```python
from sklearn.ensemble import AdaBoostClassifier
abc =AdaBoostClassifier(algorithm='SAMME',n_estimators=1,learning_rate=0.1)
abc.fit(x,y)
plot_decision_boundry(abc)
```

> **10 weak learners**

```python

abc_2 =AdaBoostClassifier(algorithm='SAMME.R',n_estimators=10,learning_rate=0.1)
abc_2.fit(x,y)
plot_decision_boundry(abc_2)
```

> **50 weak learners**

```python
abc_3 =AdaBoostClassifier(algorithm='SAMME.R',n_estimators=50,learning_rate=0.1)
abc_3.fit(x,y)
plot_decision_boundry(abc_3)
```

> **500 weak learners**

```python
abc_4 =AdaBoostClassifier(algorithm='SAMME.R',n_estimators=500,learning_rate=0.1)
abc_4.fit(x,y)
plot_decision_boundry(abc_4)
```
