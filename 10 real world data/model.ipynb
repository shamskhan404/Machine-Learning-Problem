> **Problem :** find a real world dataset.

```python
import pandas as pd
data = pd.read_csv('smartphones.csv')
data.head()
```

> **x features (price, av_rat, cores, speed)  & y target : 5G or Not** 

```python
x = data.iloc[:,[2,3,6,7]]
x.head()
```

```python
y = data.iloc[:,4]
y.head()
```

```python
x.info()
```

```python
y.info()
```

> preprocessing 

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer()
x_trf = si.fit_transform(x)
```

```python
si_2 = SimpleImputer(strategy='most_frequent')
y_trf = si_2.fit_transform((y.values).reshape(-1,1))
```

> model train

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_trf,y_trf,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

> model test

```python
y_pred = dtc.predict(x_test)
print('accuracy :',accuracy_score(y_test,y_pred))
```
